# Question 4 (b)

## Step1: Seperating two groups

```python
dfm=df1[df1['sex']=='Male']
```

```python
dff=df1[df1['sex']=='Female']
```

## Step2: Calculating the paired data difference

```python
list_m=[]
list_f=[]
list_m=dfm.values.tolist()
list_f=dff.values.tolist()
list_d=[]
```

```python
bool(len(list_m)==len(list_f))
```

```python
list_m[0]
```

```python
for r in range(0,len(list_m)):
    list_d.append(list_m[r][-1]-list_f[r][-1])
```

## Step3: Testing the assumptions

```python
import scipy.stats as stats
```

```python
statistic, pvalue = stats.kstest(list_d, 'norm')
print('Statistic: %.3f, p-value: %.3f' % (statistic, pvalue))
```

```python
levene=stats.levene(dfm.iloc[:,-1].values.tolist(),dff.iloc[:,-1].values.tolist())
print(levene.pvalue)
```

```python
plt.hist(list_d,bins=10)
```

```python
plt.hist(dfm.iloc[:,-1].values.tolist(),bins=10)
```

```python
plt.hist(dff.iloc[:,-1].values.tolist(),bins=10)
```

```python
statistic, pvalue = stats.kstest(dfm.iloc[:,-1].values.tolist(),dff.iloc[:,-1].values.tolist())
print('Statistic: %.3f, p-value: %.3f' % (statistic, pvalue))
```

## Step4: Testing the sample

```python
k=stats.ttest_1samp(list_d,0)
```

```python
k.pvalue
```

```python
stats.mannwhitneyu(dfm.iloc[:,-1].values.tolist(),dff.iloc[:,-1].values.tolist(), alternative = 'greater' )
```
